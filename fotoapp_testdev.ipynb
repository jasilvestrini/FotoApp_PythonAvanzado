```python
from PIL import Image, ImageOps #Importamos la función Image del paquete PIL
import time
import fotoapp as fapp
```

```python
def mostrar_menu():
    '''
    Muestra el menú de opciones al usuario
    '''
    print("\n---MENU PRINCIPAL PHOTO APP---")
    print("1. Ajustar Contraste")
    print("2. Aplicar Filtro")
    print("3. Dibujar")
    print("4. Guardar Imagen y Salir")
    print("5. Salir sin Guardar Imagen")
    print("===========================")
```

```python
def imprimir_diccionario(diccionario):
    for clave, valor in diccionario.items():
        print(f"{clave}: {valor}")
```

```python
# Función principal que maneja la interacción con el usuario

def main():
    from PIL import Image #Importamos la función Image del paquete PIL
    import time

    print ("Bienvenido a PhotoApp!!")

    while True:

      ruta_imagen = input("Ingrese la ruta de la imagen: ")
      redes_sociales = {"1":"Youtube","2": "Instagram","3": "Twitter", "4": "Facebook"}

      while True:
        print ("\n-REDES SOCIALES-")
        imprimir_diccionario(redes_sociales)
        red_social = input("en que red social la vas a publicar??: ")
        if red_social in redes_sociales:
          break
        else:
          print ("\n\nOpcion no valida!!!!")
      img = fapp.resize_imagen(ruta_imagen, redes_sociales[red_social])
      if not isinstance(img, str):
        break
      else:
        print(f"\nError: {img}\n")

    while True:
        mostrar_menu()  # Mostrar el menú de opciones
        time.sleep(.5)
        opcion = input("Seleccione una opción: ")  # Solicitar la opción del usuario

        if opcion == '1':
            img = fapp.ajuste_contraste(img)
            time.sleep(.5)
            pass

        elif opcion == '2':
          # Aplicar Filtro
          filtros_disponibles = {
            "1": "BLUR",
            "2": "CONTOUR",
            "3": "DETAIL",
            "4": "EDGE_ENHANCE",
            "5": "EDGE_ENHANCE_MORE",
            "6": "EMBOSS",
            "7": "FIND_EDGES",
            "8": "SMOOTH",
            "9": "SHARPEN"            }
          while True:
            print ("\n-FILTROS DISPONIBLES-")
            imprimir_diccionario(filtros_disponibles)
            filtro = input("Ingrese el filtro deseado: ")
            if filtro in filtros_disponibles:
              break
            else:
              print ("\n\nOpcion no valida!!!!")
          time.sleep(.5)

          img = fapp.aplicar_filtros(img, filtros_disponibles[filtro])

        elif opcion == '3':
            img = fapp.dibujar(img)
            pass

        elif opcion == '4':
            # Guardar y Salir del programa
            while True:
              ruta_imagen = input("Ingrese la ruta y el nombre de donde quiere guardar su imagen: ")
              try:
                img.save(ruta_imagen)
              except:
                print ("Error en la ruta - no existe, intente nuevamente!!")
                continue
              else:
                print("¡Imagen guardada con éxito!")
                break
            print("Saliendo de Photo App. ¡Hasta luego!")
            break

        elif opcion == '5':
            # Salir sin Guardar
            print("Saliendo de Photo App. ¡Hasta luego!")
            break

        else:
            # Opción no válida
            print("Opción no válida. Intente nuevamente.")
```

```python
# Punto de entrada del programa
if __name__ == "__main__":
    main()
```
